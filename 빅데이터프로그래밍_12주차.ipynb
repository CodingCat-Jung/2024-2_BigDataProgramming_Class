```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('기온.csv', encoding = 'cp949')
# df.info()
df.describe()
df2 = df.fillna(method = 'ffill')
# df2.info()
df2.rename(columns = {'평균기온(°C)':'avg_temp'},inplace=True)
df2.rename(columns = {'최저기온(°C)':'min_temp'},inplace=True)
df2.rename(columns = {'최고기온(°C)':'max_temp'},inplace=True)

plt.rc('font', family = 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title("서울시 2022년도 기온 변화")
plt.xlabel("일")
plt.ylabel("기온")
plt.plot(range(1, 366), df2['avg_temp'], label = '평균기온', c='y')
plt.plot(range(1, 366), df2['min_temp'], label = '최저기온', c='b')
plt.plot(range(1, 366), df2['max_temp'], label = '최고기온', c='r')
plt.legend()
plt.show()

# 3월 데이터만 추출해서 시각화
df2['일시'] = pd.to_datetime(df2['일시'], format='%Y-%m-%d')
df2['일시']

df3 = df2.set_index('일시')

df_march = df3[pd.DatetimeIndex(df3.index).month == 3]
df_march

plt.title("서울시 2022년도 3월 기온 변화")
plt.xlabel("3월")
plt.ylabel("기온")

plt.plot(range(1, 32), df_march['avg_temp'], label = '평균기온', c='y')
plt.plot(range(1, 32), df_march['min_temp'], label = '최저기온', c='b')
plt.plot(range(1, 32), df_march['max_temp'], label = '최고기온', c='r')
plt.legend()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
import numpy as np

with open("대한민국헌법.txt", "r", encoding='utf-8') as f:
    text = f.read()

# 형태소 분석 엔진 초기화
okt = Okt()

# 헌법 데이터 형태소 분석
nouns = okt.nouns(text)
print(nouns)

# 1글자 이상인 단어만 추출
words = [n for n in nouns if len(n) > 1]
print(words)

# 각 단어의 빈도수 계산
c = Counter(words)
print(c)
```
