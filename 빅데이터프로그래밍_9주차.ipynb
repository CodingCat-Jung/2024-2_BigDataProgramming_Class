```python
import numpy as np

data = np.loadtxt("chocolate_rating1.csv", delimiter=',')

print("차원: ", data.ndim)
print("모양: ", data.shape)
print("사이즈: ", data.size)

# print(data)

# 평균 평점
rating_mean = data[:, 3].mean()
print("평균 평점: ", rating_mean)

# 전체 초콜릿에서 평점이 4 이상인 초콜릿 번호 조회
high_level = data[data[:, 3] >= 4]
high_id = high_level[:, 0].astype(np.int64)
print(high_id)
print("우수 초콜릿 수: ", high_id.size)

# 우수 초콜릿의 카카오 함유량 빈도수 계산
high_kakao = high_level[:, 2]
# unique_value = np.unique(high_kakao) # 중복 제거 unique함수

unique_value, value_counts = np.unique(high_kakao, return_counts = True)


print("카카오 함유량: ", unique_value)
print("함유량별 빈도수: ", value_counts)
print("카카오 함유량 unique size: ", unique_value.size)

# 카카오 함유량 빈도수가 가장 높은 함유량
max_index = np.argmax(value_counts)
print("카카오 함유량이 가장 높은 함유량: ", unique_value[max_index])
print("우수초콜릿", high_id.size, "가지 중", value_counts[max_index], "가지의 카카오 함유량이", unique_value[max_index]*100, "%입니다.")
```
