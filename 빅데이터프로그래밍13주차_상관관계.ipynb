```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


df2020 = pd.read_csv('2020.csv', index_col = 0)
df2021 = pd.read_csv('2021.csv', index_col = 0)

df2020.describe()

# 불필요한 컬럼 삭제 -> drop()
df2020.drop(['upperwhisker', 'lowerwhisker', 'Dystopia'], axis=1, inplace=True)
df2021.drop(['upperwhisker', 'lowerwhisker', 'Dystopia'], axis=1, inplace=True)
df2020.head()

# 1. 국가의 행복지수와 경제적 성장율(GDP), 법적인 자유도(Freedom) 간의 강한 상관관계가 있는지
df1 = df2020.copy()
df1.drop(['Social support', 'Healthy life expectancy', 'Generosity'], axis=1, inplace=True)
df1.head()

numeric_df1 = df1.select_dtypes(include = {float, int}) # 수치 데이터만
coef = numeric_df1.corr(method='pearson')
print(coef)

plt.figure(figsize=(10, 6))
# sns.heatmap(coef, annot = True, cmap = 'Blues')

df2020.head()

# 2. 관용도, 사회적 후원, 행복
x1 = df2020[['Generosity', 'Social support', 'Happiness score']].copy()
x2 = df2021[['Generosity', 'Social support', 'Happiness score']].copy()

coef3 = x1.corr()
plt.figure(figsize=(10, 6))
# sns.heatmap(coef3, annot = True)

numeric_df2020 = df2020.select_dtypes(include = {float, int})

coef4 = numeric_df2020.corr()
plt.figure(figsize=(11, 6))

mask = np.zeros_like(coef4, dtype = bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(coef4, vmin = -1, vmax = 1, annot = True, cmap = 'RdYlBu_r', mask = mask)
```
