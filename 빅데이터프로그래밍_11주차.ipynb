```python
import pandas as pd
from datetime import datetime, date, time

file_list = ['2009_hr.xlsx', '2010_hr.xlsx', '2011_hr.xlsx', '2012_hr.xlsx', '2013_hr.xlsx', '2014_hr.xlsx', '2015_hr.xlsx', '2016_hr.xlsx']

df = pd.DataFrame()

for f in file_list:
  data = pd.read_excel(f)
  df = pd.concat([df, data])

# df.info()
# df.describe()

df2 = df.reset_index(drop=True)
# df2.info()
df2.head()

df['Date'] = pd.to_datetime(df['날짜'], format = '%Y%m%d') + pd.to_timedelta(df['시간'].astype(int), unit = 'h')

df.set_index(df['Date'], inplace = True)

df1 = df[['수온(℃)', '수위(el.m)', 'EC(㎲/㎝)']]
df1.columns = ['temp', 'level', 'EC']

df1.isnull().sum

df1.to_csv('water.csv', encoding= 'cp949')

```
